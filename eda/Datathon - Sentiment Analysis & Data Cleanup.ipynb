```python
import json
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
data = [json.loads(line) for line in open("concatenated_abridged.jsonl", 'r', encoding='utf-8')]
df = pd.DataFrame(data)
```

```python
df['scores'] = df.full_text.apply(lambda review: sid.polarity_scores(review))
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c > 0 else ('neut' if c == 0 else 'neg'))
```

```python
user = df.user.apply(pd.Series)
user
```

```python
user = user[['id', 
             'name', 
             'screen_name', 
             'location', 
             'description', 
             'followers_count', 
             'friends_count', 
             'listed_count',
             'created_at',
             'favourites_count',
             'verified',
             'statuses_count',
             'lang'
            ]]
```

```python
df = df[['created_at',
         'id',
         'full_text',
         'entities',
         'retweeted_status',
         'retweet_count',
         'favorite_count',
         'lang',
         'scores',
         'compound',
         'comp_score']]
```

```python
my_data = pd.concat([df, user], axis=1)
```
